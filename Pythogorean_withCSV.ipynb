
# Pythogorean triplets Visualised


Simple visualization of Pythogorean triplets.

Here we find the triplets upto the specified number, and then save it in a .csv.
Later, use the  csv to visualize em.



```python
import math
import pandas as pd

userinput = int(input("Upto what number do you want the Pythagorean Triplets ?"))

A = []
B = []
C = []

def pythagorean_triplet(n):
    for b in range(n):
        for a in range(1, b):
            c = math.sqrt( a * a + b * b)
            if c % 1 == 0:
                print(a, b, int(c))
                A.append(a)
                B.append(b)
                C.append(c)
    final = pd.DataFrame(
    {'a': A,
     'b': B,
     'c': C
    })
    final.to_csv('out.csv')
    #final.to_csv('out.csv',index=False)
    
                
            
pythagorean_triplet(userinput)
```

```python
df = pd.read_csv('out.csv')
```

```python
x = df['a'].values.tolist()
y = df['b'].values.tolist()
z = df['c'].values.tolist()
```

```python
r, w = df.shape
print(r,w)
x1 = []
for a in range(r):
    x1.append(a)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()


ax.plot(x1,x)
ax.plot(x1,y)
ax.plot(x1,z)


ax.set_title('Pythagorean triplets visualization')
ax.legend(['Side 1','hypotenuse','Side 2'])
#ax.xaxis.set_label_text('Angle Θ')
ax.yaxis.set_label_text('Length of side')


plt.show()
```
